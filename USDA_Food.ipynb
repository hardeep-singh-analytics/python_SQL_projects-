```python
import pandas as pd
```

```python
# Importing data from usda_food.txt into pandas DataFrame using pd.read_json function.
```

```python
usda_food_data = pd.read_json('usda_food.txt')
```

```python
usda_food_data
```

```python
#Task : Extract food_item (description), its food_category(group) and its protein value(nutrients)
```

```python
#Step 1: Extracting food_items column
```

```python
food_items = usda_food_data['description']
```

```python
food_items
```

```python
#Step 2: Extracting food_category column
```

```python
food_category = usda_food_data['group']
```

```python
food_category
```

```python
#Step 3: Extracting nutrients column
```

```python
nutrients = usda_food_data['nutrients']
```

```python
nutrients
```

```python
#Step 4: iterating over all the extracted columns such that we end up having a below dictionary having food_item,
# food_category and protein_value as keys and values contains respective values associated with each key.
```

```python
food_items_food_group_protein_dict = {'food_items': [], 'food_category': [], 'protein_values': []}
```

```python
nutrients_row = 0
for item, group in zip(food_items, food_category):
    food_item = item
    food_group = group
    protein_value = None # all items may not be having description = protein
    for nutrient in nutrients[nutrients_row]:
        if nutrient['description'] == 'Protein':
            protein_value = nutrient['value']
            break # if protein found then no need to go further
    if protein_value != None: # append to various list in the dict only if the nutrients row had protein
        food_items_food_group_protein_dict['food_items'].append(food_item)
        food_items_food_group_protein_dict['food_category'].append(food_group)
        food_items_food_group_protein_dict['protein_values'].append(protein_value)
    nutrients_row += 1 # This will ensure that in the next iteration, the 2nd row of nutrients will be picked
```

```python
item_group_protein_data = pd.DataFrame(food_items_food_group_protein_dict)
```
